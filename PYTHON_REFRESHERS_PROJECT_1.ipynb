```python
def del_char(s,c):
    l = len(c)
    if c in s and l ==1 :
        output = s.replace(c,"")
    else:
        output = s
    return output
s = input("Enter the string: ")
c = input("Enter charecter to be delete: ")
print(del_char(s,c))
```

```python
def shuffle(l1,l2):
    c = []
    if len(l1)!=0 and len(l2)!=0:
        for i in range (min(len(l1), len(l2))):
            c.extend([l1[i],l2[i]])
        c.extend(l1[i+1:] or l2[i+1:])
    else:
        c.extend(l1[0:] or l2[0:])
    return (c)
l1 = [0,2,4]
l2 = [1,3,5]
shuffle(l1,l2)
```

```python
def expanding(L):
    for i in range(len(L)-2):
        if(L[i]>=L[i+1]):
            return False
    return True    
L = [-1,2,4,7,8]
L1 = [1,3,7,2,-3]
```
